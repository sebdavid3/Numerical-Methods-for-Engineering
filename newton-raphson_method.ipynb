```python
#Graphic x3-x-2=0
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3-x-2

def df(x):
  return 3*x**2 - 1


x = np.linspace(-10, 10, 400)
y = f(x)
plt.plot(x, y, label='f(x) = x^3-x-2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gr√°fica de f(x) = x^2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x_values =[]
iteracion= []
error_values = []
x_values.append(5)
iteracion.append(1)
error_values.append(100)

aprox_error = 100
x_old = 5
while(aprox_error>1):
  iteracion.append(len(x_values)+1)
  x_new = x_old - f(x_old)/df(x_old) #Formula de newton
  x_values.append(x_new)
  aprox_error = abs(x_new-x_old)/x_new * 100
  error_values.append(aprox_error)
  x_old = x_new

print(x_new)

#plot x_values
plt.plot(iteracion, x_values, label='x_values')
plt.xlabel('iteracion')
plt.ylabel('x')
plt.title('Raices de x')
plt.legend()
plt.grid(True)
plt.show()
plt
```

```python
#plot error_values
plt.plot(iteracion, error_values, label='error_values')
plt.xlabel('iteracion')
plt.ylabel('error')
plt.title('Gr√°fica de error')
plt.legend()
plt.grid(True)
plt.show()
```
